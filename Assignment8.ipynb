```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
dataset = sns.load_dataset('iris')

# Create a bar plot with all bars colored blue
sns.barplot(x='species', y='sepal_length', data=dataset, color="blue")
plt.title("Average Sepal Length by Species")
plt.show()
```

```python
sns.rugplot(dataset['sepal_length'])
```

```python
sns.jointplot(x = dataset['petal_length'], y = dataset['petal_width'], kind =
'scatter')
sns.jointplot(x = dataset['petal_length'], y = dataset['petal_width'], kind = 'hex')
```

```python
sns.countplot(x='species', data=dataset)
```

```python
sns.boxplot(x='species', y='sepal_length', data=dataset)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.show()
```

```python
sns.stripplot(x='species', y='petal_length', data=dataset, jitter=True)
```

```python
sns.stripplot(x='species', y='petal_length', data=dataset, jitter=False)
```

```python
sns.swarmplot(x='species', y='petal_length', data=dataset)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
dataset_numeric = dataset.drop(columns='species')
correlation_matrix = dataset_numeric.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Heat Map of Feature Correlations")
plt.show()
```
