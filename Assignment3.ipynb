```python
#plots

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset = sns.load_dataset('iris')
dataset.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['sepal_length'], kde=True, bins=20, color="blue")
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
sns.distplot(x = dataset['sepal_length'], bins = 10, kde=False)
```

```python
#jointplot
sns.jointplot(x = dataset['petal_length'], y = dataset['petal_width'], kind =
'scatter')
sns.jointplot(x = dataset['petal_length'], y = dataset['petal_width'], kind = 'hex')
```

```python
sns.rugplot(dataset['sepal_length'])
```
