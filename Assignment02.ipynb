```python
import pandas as pd
import numpy as np

missing_values = ["Na", "na"]
df = pd.read_csv("StudentPerformance.csv", na_values=missing_values)
num_bins=5
df['math_binned'] = pd.cut(df['math score'], bins=num_bins, labels=False)

print(df[['math score', 'math_binned']].head())
```
